**Conda environment: conda_sceasy**

```r
library(Seurat)
library(sceasy)
library(reticulate)
```

```r
sceasy::convertFormat("/lustre/BIF/nobackup/perez070/scRNA-seq_PLT1-WOX5/output/3-2_anndata_obj/MINI-EX_adata_0h.h5ad", from="anndata", to="seurat",
                       outFile='/lustre/BIF/nobackup/perez070/scRNA-seq_PLT1-WOX5/output/3-1_models/MINI-EX_0h.rds')
data = readRDS('/lustre/BIF/nobackup/perez070/scRNA-seq_PLT1-WOX5/output/3-1_models/MINI-EX_0h.rds')

str(data)

#Write expression matrix
expression.matrix <- as.data.frame(as.matrix(GetAssayData(object = data, assay = "RNA", layer = "counts")))
write.table(expression.matrix, "/lustre/BIF/nobackup/perez070/scRNA-seq_PLT1-WOX5/output/4-MINI-EX/dataset1_matrix.tsv", sep='\t', quote = FALSE)

#Find markers
ath_markers_all <- FindAllMarkers(data, only.pos=TRUE, group.by="number_final_annotation_MINI-EX", slot="counts")
write.table(ath_markers_all, "/lustre/BIF/nobackup/perez070/scRNA-seq_PLT1-WOX5/output/4-MINI-EX/dataset1_allMarkers.tsv", sep = "\t", quote=FALSE)


data <- ScaleData(data)


DefaultAssay(data) <- "RNA"
library(tidyverse)

ath_markers_all %>%
    group_by(cluster) %>%
    dplyr::filter(avg_log2FC > 1) %>%
    slice_head(n = 2) %>%
    ungroup() -> top2


plot <- DoHeatmap(object = data, features = top2$gene, group.by = "number_final_annotation_MINI-EX")
ggplot2::ggsave(filename = "feature_0h_.png", plot = plot, width=16, height=16) # can add additional parameters similar to png 


object <-FetchData(data, vars = 'number_final_annotation_MINI-EX')
write.table(object,"/lustre/BIF/nobackup/perez070/scRNA-seq_PLT1-WOX5/output/4-MINI-EX/dataset1_cells2clusters.tsv", sep='\t',quote=FALSE,col.names = FALSE) 
```
